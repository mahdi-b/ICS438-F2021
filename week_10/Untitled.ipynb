```python
import pandas as pd
data = pd.DataFrame(iris["data"])
data.columns = iris.feature_names
data.head()
```

```python
data["species"] = iris.target
data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.pairplot(data, hue='species')
```

```python
from sklearn.decomposition import PCA
```

```python
pca =  PCA(n_components=2)
pca
```

```python
dir(pca)
```

```python
pca.fit(iris.data)
```

```python
dir(pca)
```

```python
X_pca = pca.transform(iris.data)
```

```python
dir(pca)
```

```python
pca.components_
```

```python
data["first_comp"] = X_pca[:, 0]
data.head()
```

```python
data["second_comp"] = X_pca[:, 1]
data.head()
```

```python
sns.lmplot("first_comp", "second_comp", data, fit_reg=False)
```

```python
sns.lmplot("first_comp", "second_comp", data, fit_reg=False, hue='species')
```
