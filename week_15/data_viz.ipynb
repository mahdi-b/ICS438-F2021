### Plot Legend

- Adding legends is critical for identifying plot elements
- The easiest way to add a legend is to pass a label when adding each elementand calling the legend() method of the plt object
  - The label can be a `latex` equation
- The legend is placed on the plot in a way that interferes the least with the plot
  - You can change the location of the legend manually

````python
z = 1.9**x 
plt.plot(x,y, color='blue', label="$x^2$")
plt.plot(x,z, color='#fd380d', linewidth=6, label="$1.9**x$")
plt.plot(x, x**3, color='k', linewidth=6, label="x^3")
plt.legend()
plt.xlabel("The x values go here")
plt.ylabel("The y values go here")
````

````python
z = 1.9**x 
plt.plot(x,z, color='#fd380d', linewidth=2, label="$1.9**x$", alpha=0.3)
plt.plot(x, x**3, color='k', linewidth=2, label="x^3", alpha=0.3)
plt.plot(x,y, color='blue', linewidth=3, label="$x^2$", alpha=0.3)

plt.legend()
plt.xlabel("The x values go here")
plt.ylabel("The y values go here")
````

### Subplot Arrays

* Creating an array of subplots is a common task. `matplotlib` includes a function that generates the figure and all the plots using a single function
* The plots car share the same x-axis or y-axis, 
* As with any other figures, subplots have many other options to control the spacing padding, figure size etc... 


````python
fig, axes = plt.subplots(2,3, figsize=(14,6))
print(axes)
print("*" * 20)
axes[0,0].plot(np.random.randn(50).cumsum(),  'k--')
axes[1,2].plot(np.random.randn(50).cumsum(),  'r')

# note that we cannot use plt.figure(figsize=(12,10))

````

````python
fig, axes = plt.subplots(2,3, sharex=True, sharey=True, figsize=(14,6))

axes[0,0].plot(np.random.randn(50).cumsum(),  'k--')
axes[1,2].plot(np.random.randn(50).cumsum(),  'r')

plt.subplots_adjust(wspace=0, hspace=0)
````
